## Add New product / Edit specifics of a product in Inventory

``````````````python
stop = True
while(stop):
    prod_ID = input("Enter the Product ID: ")
    prod_name = str(input("Enter the product name: "))
    qty = int(input("Enter the quantity of the product: "))
    price = int(input("Enter the price of the product: "))
    category = input("Enter the category of the product: ")
    exp = input("Enter the Expiry date of the product: ")

    products[prod_ID] = {'name':prod_name,'quantity':qty,'price':price,'category':category,"expiry date":exp}

    user_input = input("Would u like to add/edit a product (y/n): ").lower()
    if(user_input == 'n'):
        stop = False 
        

js = json.dumps(products)

fd = open("update.json",'w')
fd.write(js) 
fd.close()
``````````````

## Delete a product from Inventory

``````````````python
pid = list(products.keys())
    
print(*pid)
``````````````

``````````````python
stop = True

while(stop):
    pro_ID = input("Enter the product ID: ")
    
    if(pro_ID in pid):
        del products[pro_ID]
        print("-----Deleted Successfully-----")
        
    else:
        print("Entered product ID is not present in the Inventory")
    
    
    user_input = input("Would u like to delete another product - (y/n): ").lower()
    
    if(user_input == 'n'):
        stop = False 

``````````````

### One product

``````````````python
# Creating empty dictionary for sale
sale = {}
``````````````

``````````````python
# Code to buy only one product from inventory

ui_name = str(input("Enter your name: "))
ui_prod = input("Enter the product ID: ")

if ui_prod in pid:
    
    ui_qty = int(input("Enter the quantity: "))
    qty = products[ui_prod]['quantity']
    
    if(ui_qty < qty):                 
        qty = qty - ui_qty
        
    elif(ui_qty > qty):
        print("We have " + str(qty) + " items in the stock" )
        
        if(qty != 0):
            user = input("Do you want to buy all of them - (y/n): ").lower()
            if(user == 'y'):
                qty = 0
        else:
            print("Sorry, Do visit again")
            ui_qty = 0
        
else:
    print("We don't have the product with Id: ",ui_prod)
``````````````

``````````````python
sale[len(sale) + 1] = {'Customer name':ui_name,'Product' : ui_prod, 'Quantity': ui_qty, 'Amount': products[ui_prod]['price'] * ui_qty}
fd = open("sales.json",'w')
sales = json.dumps(sale)
fd.write(sales)
fd.close()
``````````````

``````````````python
# Bill processing

from datetime import datetime
print("==================================================")
print("Name of the customer: ",ui_name)
print("Date & Time: ",datetime.now())
print("==================================================")
print("Product name: ",products[ui_prod]['name'])
print("Quantity: ",ui_qty)
print("Price: ",products[ui_prod]['price'])
print("--------------------------------------------------")
print("Billing Amount: ",products[ui_prod]['price'] * ui_qty) 
print("==================================================")
print("                 THANK YOU :-)\n              PLEASE VISIT AGAIN ")
``````````````

### Multiple products

``````````````python
# Code to buy multiple products from inventory

stop = True
items = []
quan = []
g_total = 0

ui_name = input("Enter your name: ")

while(stop):
    
    ui_prod = input("Enter the product ID: ")
   
    if(ui_prod in pid):
        ui_qty = int(input("Enter the quantity: "))
        
        if(ui_qty < products[ui_prod]['quantity']):
            print("Product name: ",products[ui_prod]['name'])
            print("Price: ",products[ui_prod]['price'])
            print("----------------------------------------")
            print("Amount: ",products[ui_prod]['price'] * ui_qty)
            print("----------------------------------------")
        
            products[ui_prod]['quantity'] = products[ui_prod]['quantity'] - ui_qty
            
            total = ui_qty * int(products[ui_prod]['price'])
            g_total = g_total + total
            
            
        else:
            print("Sorry, We're Out of Stock")
        items.append(products[ui_prod]['name'])
        quan.append(ui_qty)
       
        ui = input("Would u like to buy another item ??? \nType 'c' to continue or 'q' to quit :").lower()
        if(ui == 'q'):
            stop = False 
        
    else:
        print("We don't have that product with Id:",ui_prod)
``````````````

``````````````python
# Bill processing
print("========================================")
print("Name of the customer: ",ui_name)
print("Date & Time: ",datetime.now())
print("Items purchased: ",items)
print("Quantity: ",quan)
print("----------------------------------------")    
print("BILLING AMOUNT : ",g_total)
print("========================================")    
print("            THANK YOU :-)\n         PLEASE VISIT AGAIN ")
``````````````

``````````````python
sale[len(sale) + 1] = {'Customer name':ui_name,'Product' : items, 'Quantity': quan, 'Amount': g_total}
sales = json.dumps(sale)
fd = open("sales.json",'w')
fd.write(sales)
fd.close()
``````````````
